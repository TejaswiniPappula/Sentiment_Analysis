# preprocessing

```python
feedback = df['reviewText']
feedback[:10]
```

```python
processed = feedback.str.replace(r'[0-9]', '')
processed = processed.str.replace(r'_', '')
processed = processed.str.lower()
processed
```

# stopwords removal

```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
nlp = spacy.load('en_core_web_sm')
stopwords = list(STOP_WORDS)
```

```python
stopwords
```

```python
processed = processed.apply(lambda x : ' '.join(term for term in x.split() if term not in stopwords))
```

```python
processed
```

# removal of punctuations

```python
import string

def remove_punctuation(txt):
    txt_nopunct = "".join([c for c in txt if c not in string.punctuation])
    return txt_nopunct

processed = processed.apply(lambda x:remove_punctuation(x))
```

```python
processed[:10]
```

# lemmetization and tokenization 

```python
all_words = []
for sentence in processed:
    doc = nlp(sentence)
    for term in doc:
        if(term.lemma_ != "-PRON-"):
            all_words.append(term.lemma_)
```

```python
all_words
```

```python
from collections import Counter
word_freq = Counter(all_words)
print(word_freq)
```

```python
word_features  = list(word_freq.keys())[:5000]
word_features
```
